```python
new_lifeexp = pd.DataFrame(new_lifeexp, columns =['Country', 'Years'])
new_lifeexp
```

```python
new_lifeexp5 = new_lifeexp.set_index('Country')
```

```python
new_lifeexp5.plot.bar()
plt.title('Life expectancy in 2015')
plt.xlabel('Country')
plt.ylabel('Years old')
```

```python
new_lifeexp5.plot.bar(y='Years', width=0.3, color='lime')
```

## Animated scatterplot
*France, Germany, Sweden*

```python
fert = pd.read_csv('../../data/gapminder_total_fertility.csv', index_col=0)
```

```python
life = pd.read_excel('../../data/gapminder_lifeexpectancy.xlsx', index_col=0)
```

```python
fert.head(3)
```

```python
life.head(3)
```

```python
fert.columns # gives the years as data type 'strings'
```

```python
fert.columns = fert.columns.astype(int) # transformation from strings to integers
fert.columns
```

```python
fert.index
```

```python
fert.index.name = 'country'
```

```python
fert = fert.reset_index()
```

```python
fert = fert.melt(id_vars='country', var_name='year', value_name='fertility_rate')
```

```python
fert
```

```python
life.index.name = 'country'
```

```python
life = life.reset_index()
```

```python
life = life.melt(id_vars='country', var_name='year', value_name='life_expectancy')
```

```python
df = fert.merge(life, on=['country', 'year']) # should be not inner merge which is by default 
```

```python
df.head(5)
```

```python
fig = px.scatter(
    df,
    x='fertility_rate', 
    y='life_expectancy',
    animation_frame='year', 
    animation_group='country',
    size='fertility_rate', 
    color='country', 
    hover_name='country')
```

```python
fig.show()
```
